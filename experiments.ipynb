```python
import matplotlib.pyplot as plt
import pandas as pd
```

# CIFAR-10

```python
data_methods = {
    "method": ["permute-and-flip"] * 16 + ["exponential"] * 16,
    "l_threshold": [0.7] * 8 + [0.8] * 8 + [0.7] * 8 + [0.8] * 8,
    "eps": [0.1, 0.5, 1, 5, 10, 20, 50, 100] * 4,
    "accuracy": [
        0.69, 0.633, 0.625, 0.69, 0.63, 0.67, 0.689, 0.668,
        0.643, 0.577, 0.566, 0.629, 0.57, 0.60, 0.63, 0.61,
        0.681, 0.672, 0.655, 0.652, 0.66, 0.65, 0.694, 0.669,
        0.617, 0.612, 0.595, 0.581, 0.61, 0.59, 0.632, 0.605
    ]
}

df_methods = pd.DataFrame(data_methods)

plt.figure(figsize=(14, 10))
for method in df_methods["method"].unique():
    for threshold in df_methods["l_threshold"].unique():
        subset = df_methods[(df_methods["method"] == method) & (df_methods["l_threshold"] == threshold)]
        plt.plot(subset["eps"], subset["accuracy"], marker='o', label=f'{method} (l_threshold={threshold})')

plt.title("Accuracy vs Eps for Different Methods and l_thresholds")
plt.xlabel("Eps")
plt.ylabel("Accuracy")
plt.xscale('log')  # eps log scale 
plt.legend()
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
