## Plot

```python
def load_data(savedir):
    """
    Load our saved scores and then put them into a big matrix.
    """
    global scores, keep, upstream_scores, upstream_keep
    scores = []
    keep = []

    for path in os.listdir(savedir):
        scores.append(np.load(os.path.join(savedir, path, "scores.npy")))
        keep.append(np.load(os.path.join(savedir, path, "keep_sampled.npy")))
    scores = np.array(scores)
    keep = np.array(keep)

    upstream_scores = np.load(os.path.join(savedir, "upstream", "scores.npy"))
    upstream_keep = np.load(os.path.join(savedir, "upstream", "keep_sampled.npy"))

    return scores, keep, upstream_scores, upstream_keep
```

```python
path = "exp/resnet50_cifar10"
scores, keep, upstream_scores, upstream_keep = load_data(path)
```
